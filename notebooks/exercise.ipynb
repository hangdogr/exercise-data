```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("valakhorasani/gym-members-exercise-dataset")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(f'{path}/gym_members_exercise_tracking.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# Data Wrnagling
df = pd.concat([df, pd.get_dummies(df['Gender'], prefix='Gender')], axis=1)
df.drop(columns=['Gender'], inplace=True)

df = pd.concat([df, pd.get_dummies(df['Workout_Type'], prefix='Workout')], axis=1)
df.drop(columns=['Workout_Type'], inplace=True)
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Feature Correlation")
plt.show()
```
